# COBOL-to-Rust Interactive Preview

This notebook demonstrates loading COBOL source, preprocessing copybooks, running conversion, viewing diagnostics, and previewing generated Rust.

```python
import json
from pathlib import Path

import cobol2rust
```

```python
source_path = Path("sample.cbl")
source = source_path.read_text(encoding="utf-8") if source_path.exists() else """
IDENTIFICATION DIVISION.
PROGRAM-ID. NOTEBOOK.
DATA DIVISION.
WORKING-STORAGE SECTION.
COPY FIELDS.
PROCEDURE DIVISION.
MAIN.
    DISPLAY WS-FLAG.
    STOP RUN.
"""

copybooks = {"FIELDS.cpy": "01 WS-FLAG PIC X VALUE \"Y\".\n"}
```

```python
expanded = cobol2rust.preprocess(source, copybooks, source_format="free")
print(expanded)
```

```python
result = cobol2rust.convert_cobol(
    source,
    "ibm",
    {"source_format": "free", "copybooks": copybooks},
)

if result["ok"]:
    print(result["rust"][:1200])
else:
    diagnostics = json.loads(result["diagnostics_json"])
    diagnostics
```
